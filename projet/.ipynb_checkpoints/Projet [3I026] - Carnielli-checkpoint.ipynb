## 1. Lecture et transformation de la base TMDb

On a commencé ce projet par convertir la base de données TMDb, fournie sous le format `pkl`, vers un format plus facilement exploitable. On a ainsi sélectionné des données qui peuvent être utiles dans ce projet pour les enregistrer sous la forme d’un fichier CSV. Cette transformation est faite une seule fois et permet de récupérer plus facilement ces données par la suite sous la forme d’un dataframe `pandas`. On a choisi pour l'instant de garder la plupart des informations, même celles qui ne seront pas directement utilisées dans la suite de ce projet.

```python
# Ouverture des fichiers
with open(path + "act_v2.pkl", "rb") as file_act :
    act_pkl = pkl.load(file_act)
    
with open(path + "film_v2.pkl", "rb") as file_film :
    film_pkl = pkl.load(file_film)
    
with open(path + "crew_v2.pkl", "rb") as file_crew :
    crew_pkl = pkl.load(file_crew)
```

À partir du fichier `act_v2.pkl`, on a crée deux fichiers CSV, un premier avec des informations de chaque acteur (id, nom et genre) et un deuxième en reliant acteurs et films dans lesquels ils ont joué, contenant, dans chaque ligne, l’id d’un acteur, l’id d’un filme dans lequel il a joué, son ordre dans le film et le personnage qu’il a joué.

À partir de `film_v2.pkl`, on crée un fichier CSV contenant des informations sur les films, qui comprennent leur id dans TMDb, un booléen indiquant s'il s'agit d'un filme adulte, la langue et le titre d’origine, la date de sortie, le titre en anglais, la moyenne et le nombre de votes dans TMDb et la mesure de popularité de TMDb.

Finalement, à partir de `crew_v2.pkl`, on crée deux fichiers CSV, l’un avec les informations de chaque membre de l’équipe de production du film avec leur nom et genre et l’autre avec les relations entre ces personnes et les films, indiquant, pour chaque personne et chaque filme dans lequel elle a participé, dans quel département elle a travaillé et quel était son rôle.

```python
with open(path + "generated/actors.csv", "w", encoding="utf-8") as f_actors, \
    open(path + "generated/actorsMovies.csv", "w", encoding="utf-8") as f_actorsMovies, \
    open(path + "generated/movies.csv", "w", encoding="utf-8") as f_movies, \
    open(path + "generated/crew.csv", "w", encoding="utf-8") as f_crew, \
    open(path + "generated/crewMovies.csv", "w", encoding="utf-8") as f_crewMovies:
        
    f_actors.write("idActor;name;gender\n")
    f_actorsMovies.write("idActor;tmdbId;order;character\n")
    f_movies.write("tmdbId;adult;originalLanguage;originalTitle;releaseDate;titleTmdb;ratingTmdb;nbRatingTmdb;popularity\n")
    f_crew.write("idCrew;name;gender\n")
    f_crewMovies.write("idCrew;tmdbId;department;job\n")
    actorsAdded = set()
    crewAdded = set()

    # act_pkl, film_pkl et crew_pkl ont tous la même longueur
    for i in range(len(film_pkl)):
        f = film_pkl[i]
        tmdbId = f["id"]

        f_movies.write("{:d};{:d};{:s};{:s};{:s};{:s};{:f};{:d};{:f}\n"\
                          .format(f["id"], f["adult"], f["original_language"],\
                                  f["original_title"] if ";" not in f["original_title"] \
                                                      else  "\"" + f["original_title"] + "\"",\
                                  "" if "release_date" not in f else f["release_date"],\
                                  f["title"] if ";" not in f["title"] else  "\"" + f["title"] + "\"",\
                                  f["vote_average"],f["vote_count"],\
                                  0 if "popularity" not in f else f["popularity"]))
        for actor in act_pkl[i]:
            if actor["id"] not in actorsAdded:
                actorsAdded.add(actor["id"])
                f_actors.write("{:d};{:s};{:d}\n".format(actor["id"], actor["name"], actor["gender"]))
            f_actorsMovies.write("{:d};{:d};{:d};{:s}\n".format(actor["id"], tmdbId, actor["order"],\
                                                                actor["character"].replace(";",",")))

        for cr in crew_pkl[i]:
            if cr["id"] not in crewAdded:
                crewAdded.add(cr["id"])
                f_crew.write("{:d};{:s};{:d}\n".format(cr["id"], cr["name"], cr["gender"]))
            f_crewMovies.write("{:d};{:d};{:s};{:s}\n".format(cr["id"], tmdbId, cr["department"], cr["job"]))
```

```python
genres = ["Action", "Adventure", "Animation", "Children", "Comedy", \
          "Crime", "Documentary", "Drama", "Fantasy", "Film-Noir", "Horror", \
          "Musical", "Mystery", "Romance", "Sci-Fi", "Thriller", "War", \
          "Western"]

for g in genres:
    moviesML[g.lower()] = moviesML.apply(lambda r: 2*int(g in r["genres"]) - 1, axis=1)
moviesML.drop("genres", axis=1, inplace=True)
```

```python
dfTemp = ratingsML.groupby("movieId")["rating"].agg(["mean", "count"]).reset_index()
dfTemp.rename({"mean": "ratingML", "count": "nbRatingML"}, axis=1, inplace=True)
moviesML = pd.merge(moviesML, dfTemp, on="movieId", how="inner")
```

```python
linksML = linksML.loc[linksML["tmdbId"]!=""]
linksML["tmdbId"] = pd.to_numeric(linksML["tmdbId"])
```

```python
dfTemp = pd.merge(moviesML, linksML, on="movieId", how="inner")
movies = pd.merge(dfTemp, moviesTmdb, on="tmdbId", how="inner")
movies['releaseWeek'] = movies.releaseDate.apply(lambda x: pd.to_datetime(x).week)
movies['releaseMonth'] = movies.releaseDate.apply(lambda x: pd.to_datetime(x).month)
movies['releaseYear'] = movies.releaseDate.apply(lambda x: pd.to_datetime(x).year)
movies.drop(["imdbId", "adult", "popularity", "originalTitle", "title", "releaseDate"], axis=1, inplace=True)
movies.rename({"titleTmdb": "title"}, axis=1, inplace=True)
```

```python
dpts = np.unique(crewMoviesTmdb["department"])
for dpt in dpts:
    colName = "nbCrewDpt" + dpt.replace(" ", "").replace("&","")
    movies = movies.merge(crewMoviesTmdb[["tmdbId", "department"]][crewMoviesTmdb["department"]==dpt]\
    .groupby("tmdbId").count().rename({"department": colName}, axis=1).reset_index(), \
                 on="tmdbId", how="left")
    movies[colName].fillna(0, inplace=True)
```

## 3. Classification supervisée

### 3.1. Classification par genre

On commence les problèmes de classification par un cas simple : étant données les évaluations qu’un film a eu sur MovieLens et TMDb, peut-on savoir s’il appartient à un genre en particulier ? C’est un problème en dimension 2, ce qui permet en particulier de visualiser simplement les données. On choisit ici le genre « drame » car il est très bien représenté dans le dataset `movies`, avec environ 50 % des films classifiés dans ce genre. Cela donne ainsi une bonne quantité de films de drame pour l’apprentissage.

Comme cette classification se base sur les notes données par les utilisateurs, on trouve judicieux de ne considérer que les films qui ont été évalués par une quantité significative d’utilisateurs. On choisit ainsi uniquement les films qui ont au moins 100 notes sur MovieLens, ce qui laisse encore une quantité importante de films dans la base de données. On n’a pas fait le même avec le nombre d’utilisateurs ayant noté les films sur TMDb car cette base de données est moins utilisée et un tri similaire aurait réduit de façon plus significative la quantité de films.

Pour utiliser les classifieurs codés en cours, on crée un LabeledSet avec les données. On choisit de normaliser les données d’évaluation avec une moyenne de zéro et une variance de 1 car les perceptrons de base ont comme frontière de décision un hyperplan passant par l’origine, ce qui les rend très mauvais pour classifier des données ayant un signe, comme les notes, toujours positives.

```python
moviesUsed = movies[movies["nbRatingML"] >= 100]

columnsX = ["ratingML", "ratingTmdb"]
columnY = ["drama"]

labSet = ls.LabeledSet(len(columnsX))
labSet.x = (moviesUsed[columnsX].values - moviesUsed[columnsX].mean().values)/moviesUsed[columnsX].std().values
labSet.y = moviesUsed[columnY].values.reshape(-1, 1)
labSet.nb_examples = labSet.y.shape[0]

labSetLearn, labSetTest = ut.split(labSet, p_train = 0.85)

print("Quantité de données dans le LabeledSet : {:d}".format(labSet.nb_examples))
```

```python
print("Proportion de films du genre " + columnY[0] + " dans la base de données : {:.2f}%"\
      .format(100*(labSet.y==1).sum()/labSet.nb_examples))
```

```python
ut.plot2DSet(labSet)
```

```python
N = 100 # Nombre de tours d'apprentissage
```

```python
classif = cl.ClassifierGradientSto(len(columnsX), 1e-3)

accuracyLearn = np.zeros(N)
accuracyTest = np.zeros(N)

for i in range(N):
    accuracyLearn[i] = classif.accuracy(labSetLearn)
    accuracyTest[i] = classif.accuracy(labSetTest)
    classif.train(labSetLearn)
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
ax.grid(True)
ax.set_axisbelow(True)
ax.plot(accuracyLearn, label="Apprentissage")
ax.plot(accuracyTest, label="Test")
ax.legend(loc="lower right")
ax.set_xlabel("Itérations")
ax.set_xticks(range(0, N, int(N/10)))
ax.set_ylabel("Accuracy")
plt.show()
```

```python
print(classif.accuracy(labSetLearn), classif.accuracy(labSetTest))
```

On a ainsi une _accuracy_ d’environ 60 % pour les données d’apprentissage et de test. Cela est un bon résultat, notamment car ce classifieur se base sur très peu d’informations &mdash; les deux notes &mdash; et il est bien supérieur à ce qu’on obtiendrait, par exemple, avec un classifieur aléatoire ou par classe majoritaire.

En vue de ce résultat positif, on teste d’autres classifieurs vus en cours sur ce même problème de classification.

Perceptron par gradient stochastique kernelisé :

```python
kernel = ut.KernelPoly()
classif = cl.ClassifierGradientStoKernel(kernel.transform(labSet.x[0]).size, 1e-3, kernel)

accuracyLearn = np.zeros(N)
accuracyTest = np.zeros(N)

for i in range(N):
    accuracyLearn[i] = classif.accuracy(labSetLearn)
    accuracyTest[i] = classif.accuracy(labSetTest)
    classif.train(labSetLearn)
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
ax.grid(True)
ax.set_axisbelow(True)
ax.plot(accuracyLearn, label="Apprentissage")
ax.plot(accuracyTest, label="Test")
ax.legend(loc="lower right")
ax.set_xlabel("Itérations")
ax.set_xticks(range(0, N, int(N/10)))
ax.set_ylabel("Accuracy")
plt.show()
```

On peut se demander si, avec plus d’informations, comme par exemple le nombre d’acteurs participant du film, le nombre de membres de l’équipe de production ou la quantité d’utilisateurs de MovieLens et TMDb ayant noté le film, on peut améliorer le résultat. Ce test est fait dans les cellules suivantes pour les comédies, qui représentent environ 37 % de la base de données. On utilise un perceptron avec gradient stochastique.

```python
moviesUsed = movies[movies["nbRatingML"] >= 100]

columnsX = ["nbActors", "nbCrew", "ratingML", "nbRatingML", "ratingTmdb", "nbRatingTmdb"]
columnY = ["comedy"]

labSet = ls.LabeledSet(len(columnsX))
labSet.x = (moviesUsed[columnsX].values - moviesUsed[columnsX].mean().values)/moviesUsed[columnsX].std().values
labSet.y = moviesUsed[columnY].values.reshape(-1, 1)
labSet.nb_examples = labSet.y.shape[0]

labSetLearn, labSetTest = ut.split(labSet, p_train = 0.85)

print("Quantité de données dans le LabeledSet : {:d}".format(labSet.nb_examples))
```

```python
print("Proportion de films du genre " + columnY[0] + " dans la base de données : {:.2f}%"\
      .format(100*(labSet.y==1).sum()/labSet.nb_examples))
```

```python
N = 100 # Nombre de tours d'apprentissage
```

```python
classif = cl.ClassifierGradientSto(len(columnsX), 1e-3)

accuracyLearn = np.zeros(N)
accuracyTest = np.zeros(N)

for i in range(N):
    accuracyLearn[i] = classif.accuracy(labSetLearn)
    accuracyTest[i] = classif.accuracy(labSetTest)
    classif.train(labSetLearn)
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
ax.grid(True)
ax.set_axisbelow(True)
ax.plot(accuracyLearn, label="Apprentissage")
ax.plot(accuracyTest, label="Test")
ax.legend(loc="lower right")
ax.set_xlabel("Itérations")
ax.set_xticks(range(0, N, int(N/10)))
ax.set_ylabel("Accuracy")
plt.show()
```

```python
print(classif.accuracy(labSetLearn), classif.accuracy(labSetTest))
```

On remarque que les valeurs d’_accuracy_ obtenues sont assez similaires à celles du cas précédent. Cependant, en comparant ce classifieur avec un classifieur par classe majoritaire, on remarque qu’il n’est pas aussi bon que le précédent. En effet, dans la classification de films de drame, comme on a environ 49 % de films de drame dans le dataset, une classification par classe majoritaire classifie tout film comme non-drame, donnant une réponse correcte dans environ 51 % des cas. Dans cette classification de comédies, comme environ 37 % des films dans la base de données sont des comédies, la classification par classe majoritaire donne une réponse correcte dans 63 % des cas. Ce classifieur n’est donc pas mieux qu’une classification par classe majoritaire.

### 3.2. Classification par genre avec plus de features

La classification faite dans la partie précédente donne des résultats plutôt satisfaisants pour la classification des drames mais pas pour celle des comédies. On peut imaginer que, avec plus de _features_ sélectionnés, on peut avoir une meilleure classification. Comme dans la partie précédente, on ne prend en compte que les films ayant eu plus de 100 votes sur MovieLens.

```python
moviesUsed = movies[movies["nbRatingML"] >= 100]
```

Intuitivement, on peut imaginer que la classification d’un film peut être déterminée par les acteurs qui y jouent, certains acteurs faisant souvent des films de même genre ou similaires, ainsi que par son directeur. On essaie donc de prendre en considération les cinq principaux acteurs d’un film et son directeur pour déterminer la classification du film. On prend également en considération les _features_ précédents.

Les classifieurs codés ne permettant que l’utilisation de données numériques, il faut représenter les acteurs et directeurs par des nombres. On a choisi d’utiliser leur identifiant dans TMDb, déjà disponible dans les dataframes `actorMoviesTmdb` et `crewMoviesTmdb`.

Pour simplifier la création des LabeledSets, on crée une fonction qui transforme une ligne du dataframe `movies` en une entrée d’un LabeledSet. Pour cela, grâce à l’identifiant TMDb inclus dans la ligne, on récupère, dans le dataFrame `actorsMoviesTmdb` préalablement regroupé par film, la liste des acteurs que jouent dans ce film afin d’en extraire les cinq premiers identifiants. De même, on récupère depuis `crewMoviesTmdb` le directeur du film, s’il est présent, en ne prenant que le premier directeur s’il y en a plusieurs. Les autres entrées sont prises directement du dataframe `movies`. Cette fonction normalise aussi les données en leur soustrayant la moyenne et divisant par l’écart-type, passés en argument à la fonction.

```python
def lineToVector(line, actorsMoviesGroups, crewMoviesGroups, moyennes, ecarts):
    vetor = np.zeros(12)
    tmdbId = line["tmdbId"]
    group_act = actorsMoviesGroups[tmdbId]
    actors = actorsMoviesTmdb.loc[group_act].sort_values("order")
    for i in range(min(5, len(group_act))):
        vetor[i] = actors.iloc[i]["idActor"] 
    group_crew = crewMoviesGroups[tmdbId]
    crew = crewMoviesTmdb.loc[group_crew]
    line_director = crew[crew["job"]=="Director"]
    if line_director.shape[0] >= 1:
        vetor[5] = line_director.iloc[0]["idCrew"]
    vetor[6] = line["nbActors"]
    vetor[7] = line["nbCrew"]
    vetor[8] = line["ratingML"]
    vetor[9] = line["nbRatingML"]
    vetor[10] = line["ratingTmdb"]
    vetor[11] = line["nbRatingTmdb"]
    return (vetor - moyennes) / ecarts
```

On crée ensuite le LabeledSet afin de classifier les comédies et on le sépare en ensembles d’apprentissage et de test.

```python
#Acteurs de 1 a 5, Directeur, nbActors, nbCrew, ratingML, nbRatingML, ratingTmdb, nbRatingTmdb
lsGenre = ls.LabeledSet(12)
actorsMoviesGroups = actorsMoviesTmdb.groupby("tmdbId").groups
crewMoviesGroups = crewMoviesTmdb.groupby("tmdbId").groups

moyennes = np.zeros(12)
ecarts = np.zeros(12)

moyennes[:5] = actorsTmdb.idActor.values.mean()
ecarts[:5] = actorsTmdb.idActor.values.std()
moyennes[5] = crewTmdb.idCrew.values.mean()
ecarts[5] = crewTmdb.idCrew.values.std()
moyennes[6:] = moviesUsed[["nbActors", "nbCrew", "ratingML", "nbRatingML", "ratingTmdb", "nbRatingTmdb"]].mean()
ecarts[6:] = moviesUsed[["nbActors", "nbCrew", "ratingML", "nbRatingML", "ratingTmdb", "nbRatingTmdb"]].std()

for _, line in moviesUsed.iterrows():
    lsGenre.addExample(lineToVector(line, actorsMoviesGroups, crewMoviesGroups, moyennes, ecarts), line["comedy"])
```

```python
lsGenreLearn, lsGenreTest = ut.split(lsGenre, p_train = 0.85)
```

Les résultats obtenus sont essentiellement les mêmes que ceux de la partie précédente, avec des _accuracy_ de l’ordre de 63 %, soit l’accuracy que l’on aurait obtenu par un classifieur par classe majoritaire, puisqu’il y a environ 37 % de comédies dans la base de données.

L’origine de ces résultats décevants s’explique par la façon dont les acteurs et directeurs sont codés. En effet, en associant un numéro essentiellement arbitraire à chaque acteur ou directeur, on crée plus de structure sur les données, car on a, par exemple, une relation de distance entre eux : les acteurs auxquels on associe les numéros 1 et 2 sont plus proches entre eux que les acteurs 1 et 3. Or, il n’y a aucune raison pour que des acteurs ayant des numéros proches aient des caractéristiques en commun. Ainsi, cette façon de représenter acteurs et directeurs par des nombres n’est pas très adaptée, car on ne s’attend pas à avoir beaucoup d’information sur le numéro de l’acteur en soi. Cela explique pourquoi cette approche, bien que fondée sur une intuition raisonnable que les acteurs ont une relation avec le genre du film, ne donne pas des bons résultats.

Une solution serait de faire un _one-hot encoding_ où on rajoute dans le dataframe `movies` autant de colonnes qu’il y a d’acteurs dans la base de données et on les rempli avec 1 si l’acteur a joué dans ce film et -1 sinon, en procédant de même pour les directeurs. Cependant, la base de données `actorsTmdb` contient 165302 acteurs, une quantité qui rend cette approche prohibitive.

## 4. Régression supervisée

On cherche maintenant à estimer la note qu’un film a eu sur MovieLens à partir des informations disponibles. Pour cela, comme dans les tests précédents, on décide de se restreindre aux films ayant eu au moins 100 évaluations sur MovieLens.

```python
moviesUsed = movies[movies["nbRatingML"] >= 100]
```

La prise en compte des acteurs et directeurs faite à la partie précédente laisse à désirer à cause de l’encodage. Dans cette partie, plutôt que d’essayer d’identifier les acteurs individuellement, on cherche à calculer des caractéristiques numériques de chaque acteur et les utiliser pour obtenir des caractéristiques des films basées sur les acteurs qui y participent, avec une approche similaire pour les directeurs.

Outre les informations déjà directement disponibles sur les films sur nos bases de données, on cherche à calculer et ajouter d’autres informations qui peuvent jouer un rôle dans sa note. Ainsi, on cherche à donner une note à chaque acteur et à chaque directeur de film, qui peut être, par exemple, la note moyenne des films où ils ont participé. Ensuite, pour essayer d’estimer la note d’un filme, on utilisera aussi les notes moyennes des acteurs et directeurs y figurant.

Si cette approche est faite de façon naïve, elle peut donner lieu à une contamination. Par exemple, si un film `A` qui est dans l’ensemble de test est pris en compte dans le calcul de la note moyenne d’un acteur, et que cet acteur joue dans un autre film `B` dans l’ensemble d’apprentissage, alors la note de `A`, qu’on veut estimer, sera prise en considération indirectement pour cette estimation, car elle intervient dans le calcul de la note de l’acteur jouant dans `B`, qui est utilisé pour l’apprentissage.

Ainsi, dans le calcul des notes moyennes des acteurs, on ne peut prendre en compte que les films dans lesquels ils ont joué qui sont dans l’ensemble d’apprentissage. Pour avoir un ensemble d’apprentissage cohérent, on a décidé de séparer les données d’apprentissage et de test de façon temporelle : on utilise les films entre deux années données pour l’apprentissage et on choisit l’année suivante pour le test. Comme cela, on est certain de ne pas avoir de contamination. En plus, ce choix provient aussi de l’intuition que la performance passée d’un acteur ou directeur a une influence sur la note des films futurs desquels ils participent.

Pour choisir les années à considérer, on regarde d’abord la distribution des films par année dans le dataframe.

```python
fig, ax = plt.subplots()
ax.grid(True)
ax.set_axisbelow(True)
ax.hist(moviesUsed.releaseYear, bins=np.unique(moviesUsed.releaseYear.values))
ax.set_xlabel("Année")
ax.set_ylabel("Nombre de films")
ax.set_title("Nombre de films ayant plus de 100 évaluations par année")
print("Nombre de films en 2008 :",(moviesUsed.releaseYear==2008).sum())
```

Comme on voit dans le graphique ci-dessus, le nombre de films présents sur la base de données ayant au moins 100 évaluations sur MovieLens diminue drastiquement à partir de 2009. On a ainsi choisi comme année de test l’année 2008, pour laquelle on a 264 films dans la base de données. Pour l’apprentissage, on a choisi de prendre les 20 années précédent 2008, donc de 1988 à 2007. Bien que ce choix précis soit un peu arbitraire, il faut choisir une plage d’années suffisamment longue pour retrouver dans les films de cette plage une grande partie des acteurs et directeurs ayant participé à des films sortis dans l’année-test 2008.

On sépare alors le dataframe en deux dataframes, l’un pour les années entre 1988 et 2007 et l’autre pour l’année 2008. On fait les mêmes transformations dans ces deux dataframes pour avoir les mêmes colonnes à la fin.

Pour calculer la note moyenne des acteurs, on commence par joindre au dataframe `actorsMoviesTmdb` l’information de la note disponible sur le dataframe `movies1988To2007`, le résultat étant stocké dans le dataframe temporaire `dfTemp`. Intuitivement, on s’attend à ce que les acteurs principaux d’un film aient une influence beaucoup plus importante que les acteurs dans un second rôle ou encore des figurants. Ainsi, pour chaque film, on ne garde que les 5 principaux acteurs, en utilisant la colonne `order` du dataframe pour ne garder que les entrées avec `order ≤ 4`. On regroupe alors le dataframe `dfTemp` par les acteurs et on calcule la note moyenne et la note maximale des films pour chaque acteur.

On fusionne ensuite les dataframes `dfTemp` et `actorsMoviesTmdb` dans un nouveau dataframe temporaire, `dfTemp2`, qui contient donc désormais les notes moyennes et maximales de chaque acteur. On regroupe `dfTemp2` par film et on calcule la moyenne et le maximum des notes moyennes et maximales de chaque acteur d’ordre jusqu’à 4 dans chaque film. À la fin, on obtient donc dans `dfTemp2` les colonnes suivantes :

| Colonne | Contenu |
|:-----|:-----|
| `actorMeanMean` | Moyenne des notes moyennes des acteurs principaux de ce film |
| `actorMeanMax` | Plus grande note moyenne des acteurs principaux de ce film |
| `actorMaxMean` | Moyenne des notes maximales des acteurs principaux de ce film |
| `actorMaxMax` | Plus grande note maximale des acteurs principaux de ce film |

Ces informations sont ensuite ajoutées aux dataframes d’apprentissage et de test, `movies1988To2007` et `movies2008`. Remarquons que cela a pour effet de supprimer de `movies2008` tous les films n’ayant aucun acteur qui a joué dans des films entre 1988 et 2007. Comme l’idée utilisée ici est d’utiliser les notes précédentes des acteurs afin de prédire la note d’un film, il est en effet impossible avec cette méthode de faire une prévision pour un film dont tous les acteurs sont inconnus.

```python
movies1988To2007 = moviesUsed.loc[(movies.releaseYear >= 1988) & (movies.releaseYear <= 2007)]
movies2008 = moviesUsed.loc[movies.releaseYear == 2008]
```

```python
dfTemp = pd.merge(left = actorsMoviesTmdb.drop("character", axis = 1),
                  right = movies1988To2007[["tmdbId", "ratingML"]],
                  on = "tmdbId", how = "inner")
dfTemp = dfTemp[dfTemp.order <= 4]
dfTemp = dfTemp.groupby("idActor")["ratingML"].agg(["mean", "max"])
dfTemp.rename({"mean": "actorMean", "max": "actorMax"}, axis = 1, inplace = True)
```

```python
dfTemp2 = pd.merge(actorsMoviesTmdb.drop("character", axis=1), dfTemp.reset_index(), on = "idActor", how = "inner")
dfTemp2 = dfTemp2[dfTemp2.order <= 4].groupby("tmdbId")[["actorMean", "actorMax"]].agg(["mean", "max"])
dfTemp2.columns = ["actorMeanMean", "actorMeanMax", "actorMaxMean", "actorMaxMax"]
```

```python
movies1988To2007 = movies1988To2007.merge(dfTemp2.reset_index(), on = "tmdbId", how = "inner")
movies2008 = movies2008.merge(dfTemp2.reset_index(), on = "tmdbId", how = "inner")
```

On fait ensuite les mêmes procédures pour calculer des notes moyennes et maximales pour les directeurs à partir des films qu’ils ont fait entre 1988 et 2007 et ensuite calculer, pour chaque film, les quantités suivantes :

| Colonne | Contenu |
|:-----|:-----|
| `directorMeanMean` | Moyenne des notes moyennes des directeurs de ce film |
| `directorMeanMax` | Plus grande note moyenne des directeurs de ce film |
| `directorMaxMean` | Moyenne des notes maximales des directeurs de ce film |
| `directorMaxMax` | Plus grande note maximale des directeurs de ce film |

Ces colonnes sont aussi rajoutées dans `movies1988To2007` et `movies2008`. À la fin, l'ensemble d’apprentissage contient 4744 films et l'ensemble de test, 173.

```python
dfTemp = pd.merge(left = crewMoviesTmdb,
                  right = movies1988To2007[["tmdbId", "ratingML"]],
                  on = "tmdbId", how = "inner")
dfTemp = dfTemp[dfTemp.job == "Director"]
dfTemp = dfTemp.groupby("idCrew")["ratingML"].agg(["mean", "max"])
dfTemp.rename({"mean": "directorMean", "max": "directorMax"}, axis = 1, inplace = True)
```

```python
dfTemp2 = pd.merge(dfTemp.reset_index(), crewMoviesTmdb, on = "idCrew", how = "inner")
dfTemp2 = dfTemp2.groupby("tmdbId")[["directorMean", "directorMax"]].agg(["mean", "max"])
dfTemp2.columns = ["directorMeanMean", "directorMeanMax",
                   "directorMaxMean", "directorMaxMax"]
```

```python
movies1988To2007 = movies1988To2007.merge(dfTemp2.reset_index(), on = "tmdbId", how = "inner")
movies2008 = movies2008.merge(dfTemp2.reset_index(), on = "tmdbId", how = "inner")
print("Nombre de films dans l'ensemble d'apprentissage :", movies1988To2007.shape[0])
print("Nombre de films dans l'ensemble de test :", movies2008.shape[0])
```

On passe maintenant à l’apprentissage. La première étape est de mettre les données d’apprentissage et de test dans des LabeledSet. On choisit d’abord les colonnes qui seront utilisées pour l’apprentissage, définies dans la variable `columnsX`, ainsi que la colonne avec la valeur à estimer, dans ce cas la colonne `ratingML`, mise dans une liste dans la variable `columnsY`.

Pour l’ajout dans des LabeledSet, on a décidé d’inclure une colonne supplémentaire contenant toujours la valeur `1`. Cela a le même effet de faire un kernel simple qui permet à l’hyperplan passant par l’origine de ne pas avoir forcément la note 0. On normalise également les données utilisées pour la régression en soustrayant la moyenne et divisant par l’écart-type de chaque colonne. La normalisation doit être calculée de la même façon sur les ensembles d’apprentissage et de test, ce pourquoi nos utilisons dans les deux cas la moyenne et l’écart type des films entre 1988 et 2007.

```python
columnsX = ['action', 'adventure', 'animation', 'children', 'comedy',
            'crime', 'documentary', 'drama', 'fantasy', 'film-noir', 'horror',
            'musical', 'mystery', 'romance', 'sci-fi', 'thriller', 'war', 'western',
            'releaseWeek', 'releaseMonth', 'actorMeanMean', 'actorMeanMax', 'actorMaxMean',
            'actorMaxMax', 'directorMeanMean',
            'directorMeanMax', 'directorMaxMean', 'directorMaxMax']
columnsY = ["ratingML"]
```

```python
Ncols = len(columnsX) + 1 # 1 colonne en plus remplie de 1 pour faire un kernel simple à la main.
lsLearn = ls.LabeledSet(Ncols)

lsLearn.nb_examples = movies1988To2007.shape[0]
lsLearn.x = np.empty((lsLearn.nb_examples, Ncols))
lsLearn.y = np.empty((lsLearn.nb_examples, 1))

lsLearn.x[:, 0] = 1 # Kernel
lsLearn.x[:, 1:] = (movies1988To2007[columnsX] - movies1988To2007[columnsX].mean())/movies1988To2007[columnsX].std()
lsLearn.y[:, :] = movies1988To2007[columnsY]

lsTest = ls.LabeledSet(Ncols)

lsTest.nb_examples = movies2008.shape[0]
lsTest.x = np.empty((lsTest.nb_examples, Ncols))
lsTest.y = np.empty((lsTest.nb_examples, 1))

lsTest.x[:, 0] = 1 # Kernel
lsTest.x[:, 1:] = (movies2008[columnsX] - movies1988To2007[columnsX].mean())/movies1988To2007[columnsX].std()
lsTest.y[:, :] = movies2008[columnsY]
```

On fait d’abord une apprentissage par un perceptron avec descente de gradient stochastique. On trace l’évolution de la `loss` sur les ensembles d’apprentissage et de test au cours de l’apprentissage ainsi que les histogrammes des erreurs, donnant les écarts entre les notes réelles et les notes prédites. On rappelle que la `loss` est calculée par
$$
\text{loss} = \frac{1}{N_{\text{données}}} \sum_{\text{données}} (\text{valeur prédite} - \text{valeur réelle})^2.
$$

```python
N = 20 # Nombre de tours d'apprentissage
```

```python
classif = cl.ClassifierGradientSto(Ncols, 1e-4)

lossLearn = np.zeros(N)
lossTest = np.zeros(N)

for i in range(N):
    lossLearn[i] = classif.loss(lsLearn)
    lossTest[i] = classif.loss(lsTest)
    classif.train(lsLearn)
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
ax.grid(True)
ax.set_axisbelow(True)
ax.plot(lossLearn, label="Apprentissage")
ax.plot(lossTest, label="Test")
ax.legend(loc="upper right")
ax.set_xlabel("Itérations")
ax.set_xticks(range(0, N, int(N/10)))
ax.set_ylabel("Loss")
plt.show()
```

```python
errorLearn = np.empty(lsLearn.nb_examples)
errorTest = np.empty(lsTest.nb_examples)

for i in range(lsLearn.nb_examples):
    errorLearn[i] = classif.predict(lsLearn.x[i, :]) - lsLearn.y[i, 0]

for i in range(lsTest.nb_examples):
    errorTest[i] = classif.predict(lsTest.x[i, :]) - lsTest.y[i, 0]
```

```python
fix, ax = plt.subplots(1, 2, figsize=(15, 5))
errorMax = max(np.absolute(errorLearn).max(), np.absolute(errorTest).max()) + 0.01
_, xLimsLearn, _ = ax[0].hist(errorLearn, rwidth=0.8, bins = np.linspace(-errorMax, errorMax, 10))
_, xLimsTest, _ = ax[1].hist(errorTest, rwidth=0.8, bins = np.linspace(-errorMax, errorMax, 10))
ax[0].set_xticks((xLimsLearn[1:] + xLimsLearn[:-1])/2)
ax[1].set_xticks((xLimsTest[1:] + xLimsTest[:-1])/2)
for i in range(2):
    ax[i].grid(True)
    ax[i].set_axisbelow(True)
    ax[i].set_xlabel("Erreur")
    ax[i].set_ylabel("Nombre de films")
ax[0].set_title("Erreur dans les données d'apprentissage")
ax[1].set_title("Erreur dans les données de test");
```

```python
seuil = 0.5
print("Pourcentage de films avec erreur absolue inférieure ou égale à {:.2f} dans l'estimée de la note :".format(seuil))
print("\tEnsemble d'apprentissage : {:.2f}%".format(100 * (np.absolute(errorLearn) <= seuil).sum()/errorLearn.size))
print("\tEnsemble de test : {:.2f}%".format(100 * (np.absolute(errorTest) <= seuil).sum()/errorTest.size))
```

On remarque ainsi qu’on a une bonne régression. La `loss` décroît assez rapidement et reste à des valeurs petites, ce qui indique le vecteur `w` du perceptron a convergé. L’erreur la plus probable est, à chaque fois, d’être proche de 0. Les erreurs sont plus petites sur l’ensemble d’apprentissage. On remarque que le taux de films dont l’erreur absolue de la note prédite est inférieur à 0,5 est assez élevé, de l’ordre de 95 % sur l’ensemble d’apprentissage et 85 % sur l’ensemble de test.

```python
# Suppression du kernel simple
lsLearn.x = lsLearn.x[:, 1:]
lsTest.x = lsTest.x[:, 1:]
lsLearn.input_dimension -= 1
lsTest.input_dimension -= 1
```

```python
N = 1000
```

```python
kernel = ut.KernelPolyMultiD()
kernelSize = kernel.transform(lsLearn.x[0, :]).size
classif = cl.ClassifierGradientStoKernel(kernelSize, 1e-6, kernel)

lossLearn = np.zeros(N)
lossTest = np.zeros(N)

for i in range(N):
    lossLearn[i] = classif.loss(lsLearn)
    lossTest[i] = classif.loss(lsTest)
    classif.train(lsLearn)
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
ax.grid(True)
ax.set_axisbelow(True)
ax.plot(lossLearn, label="Apprentissage")
ax.plot(lossTest, label="Test")
ax.legend(loc="upper right")
ax.set_xlabel("Itérations")
ax.set_xticks(range(0, N, int(N/10)))
ax.set_ylabel("Loss")
plt.show()
```

```python
errorLearn = np.empty(lsLearn.nb_examples)
errorTest = np.empty(lsTest.nb_examples)

for i in range(lsLearn.nb_examples):
    errorLearn[i] = classif.predict(lsLearn.x[i, :]) - lsLearn.y[i, 0]

for i in range(lsTest.nb_examples):
    errorTest[i] = classif.predict(lsTest.x[i, :]) - lsTest.y[i, 0]
```

On remarque que la kernélisation n’a pas aidé à avoir une meilleure estimée de la note, ce qui indique que le facteur limitant de l’approche n’est pas la linéarité du perceptron utilisé précédemment mais plutôt les _features_ choisis qui ne contiennent pas plus d’information permettant d’obtenir une meilleure estimée de la note.

On a également voulu tester si la prise en compte du budget et du revenu d’un film peuvent améliorer les résultats, avec l’intuition qu’un film avec un budget plus grand ou ayant fait plus de revenu pourrait avoir une meilleure note. On refait ainsi l’expérience de classification avec les dataframes `movies1988To2007` et `movies2008` en ne gardant que ceux qui ont budget et revenu renseignés. On fait uniquement la régression avec un perceptron par descente de gradient stochastique.

```python
movies1988To2007 = movies1988To2007.dropna()
movies2008 = movies2008.dropna()
```

```python
columnsX = ['action', 'adventure', 'animation', 'children', 'comedy',
            'crime', 'documentary', 'drama', 'fantasy', 'film-noir', 'horror',
            'musical', 'mystery', 'romance', 'sci-fi', 'thriller', 'war', 'western',
            'releaseWeek', 'releaseMonth', 'actorMeanMean', 'actorMeanMax', 'actorMaxMean',
            'actorMaxMax', 'directorMeanMean',
            'directorMeanMax', 'directorMaxMean', 'directorMaxMax', 'budget', 'revenue']
columnsY = ["ratingML"]
```

```python
Ncols = len(columnsX) + 1 # 1 colonne en plus remplie de 1 pour faire un kernel simple à la main.
lsLearn = ls.LabeledSet(Ncols)

lsLearn.nb_examples = movies1988To2007.shape[0]
lsLearn.x = np.empty((lsLearn.nb_examples, Ncols))
lsLearn.y = np.empty((lsLearn.nb_examples, 1))

lsLearn.x[:, 0] = 1 # Kernel
lsLearn.x[:, 1:] = (movies1988To2007[columnsX] - movies1988To2007[columnsX].mean())/movies1988To2007[columnsX].std()
lsLearn.y[:, :] = movies1988To2007[columnsY]

lsTest = ls.LabeledSet(Ncols)

lsTest.nb_examples = movies2008.shape[0]
lsTest.x = np.empty((lsTest.nb_examples, Ncols))
lsTest.y = np.empty((lsTest.nb_examples, 1))

lsTest.x[:, 0] = 1 # Kernel
lsTest.x[:, 1:] = (movies2008[columnsX] - movies1988To2007[columnsX].mean())/movies1988To2007[columnsX].std()
lsTest.y[:, :] = movies2008[columnsY]
```

```python
N = 50 # Nombre de tours d'apprentissage
```

```python
classif = cl.ClassifierGradientSto(Ncols, 1e-4)

lossLearn = np.zeros(N)
lossTest = np.zeros(N)

for i in range(N):
    lossLearn[i] = classif.loss(lsLearn)
    lossTest[i] = classif.loss(lsTest)
    classif.train(lsLearn)
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
ax.grid(True)
ax.set_axisbelow(True)
ax.plot(lossLearn, label="Apprentissage")
ax.plot(lossTest, label="Test")
ax.legend(loc="upper right")
ax.set_xlabel("Itérations")
ax.set_xticks(range(0, N, int(N/10)))
ax.set_ylabel("Loss")
plt.show()
```

On remarque ainsi que la prise en compte du budget et du revenu n’améliore pas la qualité des résultats obtenus.

C’est intéressant aussi à regarder comment chaque colonne est prise en compte par le classifieur. La cellule ci-dessus affiche le vecteur `w` du classifieur en donnant aussi le nom des colonnes correspondantes. La première ligne, notée `w0`, correspond au coefficient multipliant la colonne remplie de `1` provenant du kernel.

```python
print("{:16s} : {:8.5f}".format("w0", classif.w[0]))
for i in range(len(columnsX)):
    print("{:16s} : {:8.5f}".format(columnsX[i], classif.w[i+1]))
```

Le coefficient `w0` vaut 3,16, ce qui correspond essentiellement à la note moyenne des films de l’ensemble d’apprentissage. C’est la valeur retournée lorsque toutes les autres entrées de `x` valent `0`, ce qui arrive si toutes les autres entrées correspondent à la moyenne de leurs valeurs.

Le genre du film semble avoir peu d’influence sur la note car les composantes de `w` correspondantes sont assez petites. Comme vu précédemment, le genre ayant plus d’influence sur la note est le drame. La semaine ou le mois de sortie ont aussi peu d’influence ; entre les deux, le mois de sortie a un poids plus important.

Ce sont notamment les indicateurs concernant les acteurs qui ont les poids le plus importants, notamment la moyenne des notes moyennes et maximales des acteurs principaux d’un film. Les directeurs ont aussi une influence sur la note, quoique d’un ordre inférieur à celle des acteurs, mais supérieur au genre du film.

Concernant le budget, il a en effet un effet négatif sur la note ! En effet, on peut imaginer que certains films ayant un grand budget n’ont pas le succès attendu. Cependant, il est intéressant à noter que les coefficients de budget et revenu sont du même ordre mais de signes opposés. Ainsi, le produit scalaire de `w` par le vecteur des _features_ calcule la différence entre revenu et budget, soit le profit du film, qui semble intuitivement une mesure plus raisonnable de succès, et donc de meilleures notes, que seul le budget.

Enfin, on remarque aussi que cette analyse des composantes de `w` n’a du sens que car toutes les données ont été normalisées.

Pour cela, on a d’abord codé un classifieur multi-classes, inclus dans le fichier `ClassifierMultiClass.py`. Ce méta-classifieur implémente une approche un-contre-tous. Il prend en argument non seulement la dimension de l’entrée et le taux d’apprentissage, comme pour d’autres perceptrons codés, mais aussi une classe de classifieur de base et une liste contenant les classes de classification.

Dans sa construction, le `ClassifierMultiClass` crée `N` classifieurs du type passé en argument, où `N` est le nombre de classes possibles. La fonction d’entraînement prend en argument un LabeledSet dans lequel l’attribut `y` a `N` colonnes, une pour chaque classe. Il entraîne ainsi chacun des classifieurs de base en leur passant en argument uniquement la colonne de `y` correspondante. Pour la prédiction, il regarde lequel des ces classifieurs rend la plus grande classification et choisit la classe correspondante. On a également dû réécrire la fonction `accuracy`, qui regarde maintenant si, dans la colonne de `y` correspondant à la classe prédite, on a bien une valeur positive.

### 6.1. Catégorisation par k moyennes

Pour la catégorisation non-supervisée, on a décidé de catégoriser les films d’un même acteur. En ajoutant cette information à `actorsTmdb` et en triant par nombre de films, on peut ainsi avoir la liste des acteurs qui ont participé à la plus grande quantité de films de la base de données.

```python
dfTemp = actorsMoviesTmdb.groupby("idActor")["order"].count().reset_index()
dfTemp = dfTemp.rename({"order": "nbMovies"}, axis = 1)
actorsNbMovies = pd.merge(actorsTmdb, dfTemp, on = "idActor", how = "inner").sort_values("nbMovies", ascending=False)
actorsNbMovies.head(20)
```

On récupère ainsi ces films du dataframe `movies` et on sélectionne des colonnes qui semblent pertinentes pour l’analyse du film, comme le genre, les notes, le nombre de notes, l’année de sortie et l’ordre de l’acteur dans le film.

```python
idActor = actorsNbMovies.loc[6122]["idActor"]
actorDB = pd.merge(movies, actorsMoviesTmdb[actorsMoviesTmdb.idActor == idActor], on = "tmdbId", how = "inner")
actorDB["isEnglish"] = 2*(actorDB.originalLanguage == "en")-1
```

```python
columns = ['action', 'adventure', 'animation', 'children', 'comedy',
           'crime', 'documentary', 'drama', 'fantasy', 'film-noir', 'horror',
           'musical', 'mystery', 'romance', 'sci-fi', 'thriller', 'war', 'western',
           'ratingML', 'nbRatingML', 'ratingTmdb', 'nbRatingTmdb',
           'releaseYear', 'order', 'isEnglish']
```

Certaines des colonnes ci-dessus ne sont pas pertinentes car elles contiennent toujours la même valeur et ne contribuent donc pas à la catégorisation des données. C’est le cas, par exemple, de la colonne `documentary` : Gérard Depardieu n’apparaît sur aucun documentaire sur cette base de données et donc cette colonne contient toujours -1. Or, comme la procédure de normalisation divise chaque colonne par la différence entre son élément maximal et son élément minimal, on aurait un division par zéro pour cette colonne, créant ainsi des valeurs `nan` dans le DataFrame. Afin de l’éviter, on détecte d’abord toutes les colonnes qui contiennent toujours la même valeur pour ne pas les prendre en considération. On identifie ces colonnes comme celles dont l’écart-type vaut 0.

```python
actorDB[columns].std()==0
```

On applique l’algorithme de k moyennes en normalisant d’abord la base de données et ensuite en calculant les k moyennes pour plusieurs valeurs différentes de `k`. On compare l’évaluation des différentes valeurs de `k` en utilisant la méthode de Dunn.

```python
columnsOK = actorDB[columns].std()!=0
columns = [c for c in columns if columnsOK[c]]
```

```python
# Normalisation:
dataKMeans = km.normalisation(actorDB[columns])
```

```python
kMax = 31
tabEval = np.full(kMax, float("inf"))

for k in range(2, kMax):
    print("Calcul de k moyennes avec k =",k)
    centres, affectation = km.kmoyennes(k, dataKMeans, 0.05, 100, verbose = False)
    tabEval[k] = km.evaluation("Dunn", dataKMeans, centres, affectation)
```

On remarque que plusieurs valeurs de `k` petits donnent essentiellement la même valeur d’évaluation, mais celle-ci croît avec `k` à partir d’environ 15 ou 20. On fait une représentation 2D des résultats à l’aide de la méthode Isomap du module `sklearn.manifold`, qui est une extension de la méthode de _multidimensional scaling_ (MDS) en utilisant la distance sur un graphe de voisins de chaque point.
```python
centres, affectation = km.kmoyennes(6, dataKMeans, 0.05, 100, verbose = False)
```

L’algorithme k moyennes arrive ainsi à bien catégoriser les points. Les quelques points un peu inconsistants qu’on observe, par exemple des poins d’une couleur trop proches des groupes de points d’une autre couleur, peuvent être expliqués par les limitations de la méthode de projection 2D.

### 6.2. Catégorisation par clustering hiérarchique

```python
columns = ['action', 'adventure', 'animation', 'children', 'comedy',
           'crime', 'documentary', 'drama', 'fantasy', 'film-noir', 'horror',
           'musical', 'mystery', 'romance', 'sci-fi', 'thriller', 'war', 'western',
           'ratingML', 'nbRatingML', 'ratingTmdb', 'nbRatingTmdb',
           'releaseYear', 'order', 'isEnglish']
```

On applique l’algorithme de clustering hiérarchique en normalisant d’abord la base de données. On a décidé d’utiliser ici la distance euclidienne.

```python
dataHierarch = ch.normalisation(actorDB[columns])
```

```python
M, dendro = ch.clustering_hierarchique_complet('euclidienne', dataHierarch, verbose = False)
```

Rappelons que, comme la distance utilisée entre deux clusters est la distance entre leurs centroïdes, les distances peuvent diminuer après une fusion, comme expliqué dans le rendu du TME7. Cela explique pourquoi certaines fusions dans le dendrogramme conduisent à des diminutions de distance.

Comme l’algorithme de clustering hiérarchique est déterministe, les résultats ne changent pas d’une exécution à l’autre, ce qui permet de commenter les résultats avec plus de détails.

On remarque que la coloration automatique a identifié sept classes : une grande classe en magenta, des classes plus petites en vert, cyan, rouge, jaune et noir, et des singletons, représentés en bleu. Pour mieux voir ce que ces clusters ont en commun, on a décidé de les isoler et voir les caractéristiques du centroïde de chaque classe.

D’après la [référence en ligne](https://docs.scipy.org/doc/scipy/reference/generated/scipy.cluster.hierarchy.dendrogram.html#scipy.cluster.hierarchy.dendrogram) de la fonction `scipy.cluster.hierarchy.dendrogram`, la coloration est calculée en prenant le seuil de 70 % de la distance maximale des fusions et en regardant toutes les classes constituées à ce seuil. Ainsi, on a crée une nouvelle fonction, `clustering_hierarchique_seuil`, qui permet d’arrêter le clustering hiérarchique à un seuil de distance et renvoie les classes formées à ce seuil (et ne trace pas de dendrogramme). On peut ainsi, comme avant, donner des caractéristiques de chaque classe en regardant la moyenne et l’écart type de chaque colonne dans chaque cluster. On rappelle que la moyenne n’est rien d’autre que le centroïde du cluster correspondant.

```python
clusters, _ = ch.clustering_hierarchique_seuil('euclidienne', dataHierarch, 0.7*M[:, 2].max(), verbose = False)
```

```python
clustersSingl = {"singletons": set()}
for cl in clusters:
    if len(clusters[cl]) == 1:
        clustersSingl["singletons"] = clustersSingl["singletons"] | clusters[cl]
    else:
        clustersSingl[cl] = clusters[cl]
print(clustersSingl)
```

```python
for i in clustersSingl:
    if i == "singletons":
        print("Singletons :")
    else:
        print("Cluster ",i,":",sep="")
    print(actorDB.iloc[list(clustersSingl[i])][columns].agg(["mean", "std"]).transpose())
    print()
```

Comme les films dans les singletons sont assez différents les uns des autres, les informations concernant la classe « singleton » ne donnent pas beaucoup d’information, elles permettent juste de vérifier, par les grands écarts-type, qu’il s’agit d’un ensemble de films assez hétérogène.

Le cluster le plus grand, représenté en magenta sur le dendrogramme, contient 53 films, tous d’action, genre par lequel Jackie Chan est connu. Ce sont les films où Jackie Chan est l’un des acteurs principaux : son ordre est de 1,15 en moyenne avec un écart-type de 2,82. La sortie de ces films a été autour de 1987 avec un écart-type d’environ 10 ans, ce qui correspond à la plage d’années où il était très actif. La majorité de ces films n’est pas en anglais mais il y a un grand mélange entre films en anglais et dans d’autres langues.

On peut faire des analyses similaires sur d’autres clusters. Le cluster 111, représenté en vert sur la figure, ne continent que deux films. En le regardant, on remarque qu’il ne contient que des documentaires faits autour de 2010, qui ont eu des évaluations plutôt bonnes mais avec peu de notes. Les deux sont en anglais. Ce sont d’ailleurs les deux uniques documentaires où figure Jackie Chan dans la base de données. Le cluster 129 se distingue pour contenir deux animations, les deux seules parmi les clusters (bien qu’il y ait une animation parmi les singletons), classées aussi comme comédies d’action pour enfants et sorties autour de 2009.

Le critère d’arrêt à 70 % de la distance maximale est arbitraire et n’a été fait ici que pour correspondre au choix par défaut de la fonction de tracé du dendrogramme. On aurait pu choisir aussi une distance plus petite, permettant d’avoir des clusters plus homogènes, ou plus grande, permettant de regrouper certains singletons en clusters.
