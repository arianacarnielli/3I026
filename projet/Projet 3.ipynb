## Import des bases de données CSV

```python
path = "data/"

movies_pd = pd.read_csv(path + "movies.csv", index_col = 0)
movies_pd.head()
```

```python
genres = ["Action", "Adventure", "Animation", "Children", "Comedy", \
          "Crime", "Documentary", "Drama", "Fantasy", "Film-Noir", "Horror", \
          "Musical", "Mystery", "Romance", "Sci-Fi", "Thriller", "War", \
          "Western", "(no genres listed)"]

for g in genres:
    movies_pd[g.lower()] = movies_pd.apply(lambda r: 2*int(g in r["genres"]) - 1, axis=1)
```

```python
movies_pd.head()
```

```python
def get_year(r):
    match = re.match("(.*)\s*\((\d+)\)", r["title"])
    if match:
        return match.group(2)
    
def get_title_without_year(r):
    match = re.match("(.*)\s*\((\d+)\)", r["title"])
    if match:
        return match.group(1)
    else:
        return r["title"]

movies_pd["Year"] = movies_pd.apply(get_year, axis = 1)
movies_pd["title"] = movies_pd.apply(get_title_without_year, axis = 1)
```

```python
movies_pd.head()
```

```python
ratings_pd = pd.read_csv(path + "ratings.csv")
ratings_pd.head()
```

```python
movies_pd["Average rating"] = ratings_pd.groupby("movieId")["rating"].mean()
movies_pd["Number of ratings"] = ratings_pd.groupby("movieId")["rating"].count()

movies_pd["Average rating"].fillna(0, inplace = True)
movies_pd["Number of ratings"].fillna(0, inplace = True)
```

## Tentativa 5

```python
ls_note_learn = ls.LabeledSet(len(filmes_pd_redux_1980_2000.columns) - 2)

ls_note_learn.x = filmes_pd_redux_1980_2000.drop(["year", "ratingMovieLens"], axis = 1).values
ls_note_learn.y = filmes_pd_redux_1980_2000.ratingMovieLens.values.reshape((-1, 1))
ls_note_learn.y = (ls_note_learn.y >= 3.5) * 2 - 1

ls_note_learn.nb_examples = ls_note_learn.y.shape[0]
```

```python
ls_note_test= ls.LabeledSet(len(filmes_pd_redux_2001.columns) - 2)

ls_note_test.x = filmes_pd_redux_2001.drop(["year", "ratingMovieLens"], axis = 1).values
ls_note_test.y = filmes_pd_redux_2001.ratingMovieLens.values.reshape((-1, 1))
ls_note_test.y = (ls_note_test.y >= 3.5) * 2 - 1

ls_note_test.nb_examples = ls_note_test.y.shape[0]
```

```python
ls_all = np.concatenate((ls_note_learn.x, ls_note_test.x))
ls_mean = ls_all.mean(0)
ls_std = ls_all.std(0)

ls_note_learn.x = (ls_note_learn.x - ls_mean)/ls_std
ls_note_test.x = (ls_note_test.x - ls_mean)/ls_std
```

```python
cl1 = cl.ClassifierPerceptron(ls_note_learn.x.shape[1], 1e-3)
N = 400
for i in range(N):
    cl1.train(ls_note_learn)
    if np.any(np.isnan(cl1.w)):
        print("w is nan")
        break
```

```python
print(cl1.accuracy(ls_note_learn), cl1.accuracy(ls_note_test))
```

```python
cl2 = cl.ClassifierGradientSto(ls_note_learn.x.shape[1], 1e-3)
N = 100
for i in range(N):
    cl1.train(ls_note_learn)
    if np.any(np.isnan(cl1.w)):
        print("w is nan")
        break
```

```python
print(cl2.accuracy(ls_note_learn), cl2.accuracy(ls_note_test))
```
