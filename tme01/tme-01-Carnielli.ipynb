Nous allons maintenant écrire les fonctions permettant de calculer des mesures ''basiques'' sur des données unidimensionnelles. Pour cela, nous allons travailler avec la structure <tt>array</tt> de numpy. Pour convertir une colonne de DataFrame en '<tt>array</tt>, voici commment faire:

### Moyenne

<font color="RED" size="+1">**[Q]**</font> **La moyenne**


Ecrire la fonction moyenne(valeurs) qui calcule la moyenne des éléments d'un tableau (en utilisant la boucle ''for'', sans utiliser les fonctions déjà implémentées dans numpy)

```python
def moyenne(valeurs):
    val=0
    for i in valeurs:
        val+=i
    return val/len(valeurs)
```

Tester la fonction moyenne et comparer vos résultats avec l'implémentation de base de python donnée ci-dessous:

```python
print("La moyenne (qualité moyenne) est : %f dollars" % prices_pd["MedQ"].mean())
print("\n")
print("La moyenne (qualité moyenne, ma fonction) est : %f dollars" % moyenne(prices_pd["MedQ"]))
```

<font color="RED" size="+1">**[Q]**</font> **Corrélations**

A l'aide de la fonction moyenne écrite précédemment, écrire une fonction <tt>correlation(x,y)</tt> qui calcule la corrélation lineaire entre deux tableaux Numpy.

```python
def correlation(x, y):
    moy_x = moyenne(x)
    moy_y = moyenne(y)
    var_x = 0
    var_y = 0
    co_var = 0
    for i in range(x.size):
        var_x += (x[i] - moy_x)**2
        var_y += (y[i] - moy_y)**2
        co_var += (x[i] - moy_x)*(y[i] - moy_y)
    e_x = np.sqrt(var_x / x.size)
    e_y = np.sqrt(var_y / y.size)
    co_var = co_var / x.size
    return co_var/(e_x * e_y)
    
```

```python
# Appliquer votre fonction avec l'instruction suivante:

print("La correlation est :%f"%correlation(prix_ca_ny["HighQ_x"].values,prix_ca_ny["HighQ_y"].values))

print(np.corrcoef(prix_ca_ny["HighQ_x"].values,prix_ca_ny["HighQ_y"].values))
```

Calculer les corrélations entre prénoms pour dégager les listes de prénoms corrélés et anti-corrélés dans le temps

```python
annees = np.unique(noms_pd["Annee"].values)
prenoms = np.unique(noms_pd["Prenoms"].values)
dict_prenom = {prenoms[i]:i for i in range(prenoms.size)}
dict_annee = {annees[i]:i for i in range(annees.size)}
tab_pren_an = np.zeros((prenoms.size, annees.size))

for index, row in noms_pd.iterrows():
    ind_prenom = dict_prenom[row["Prenoms"]]
    ind_annee = dict_annee[row["Annee"]]
    tab_pren_an[ind_prenom, ind_annee] = row["Nombre"]
    
mat_corr = np.corrcoef(tab_pren_an)
# Suppression de la diagonale pour éviter de la prendre en compte dans la recherche des correlations maximales
mat_corr = mat_corr - np.eye(prenoms.size)
abs_mat_corr = abs(mat_corr)
```

```python
print("500 correlations les plus importantes :")
for i in range(500):
    i, j = np.unravel_index(abs_mat_corr.argmax(), mat_corr.shape)
    print(prenoms[i], prenoms[j], mat_corr[i, j])
    abs_mat_corr[i, j] = 0
    abs_mat_corr[j, i] = 0
```
